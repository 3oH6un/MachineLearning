# Chapter4. 반복

## 1. while 반복문 기본

```python
# 1. while문을 이용하여 자신의 이름을 다섯번 출력해보세요.
name = "웅회"
a = 0
while a != 5:
  print(name)
  a += 1

# 2. while문을 이용해서 100보다 작은 자연수 중 홀수만 출력하세요.
a = 0
while a != 100:
  if(a % 2 == 1):
    print(a)
  a += 1
```

## 3. for 반복문(자료순회)

```python
# 좋아하는 색을 리스트로 설정하고 이를 for로 반복하여 출력하라.
colors = ['노란색', '흰색', '검은색']
for color in colors:
  print(color)
# 리스트에 1, 2, 3, 4, 5를 저장하고, for 반복으로 2, 3, 4, 5, 6을 출력하여라.
nums = [1, 2, 3, 4, 5]
for num in nums:
  print(num + 1)
```

## 4. for 횟수 반복

```python
# for와 range()를 사용하여 0부터 10까지 출력하시오.
for i in range(11):
  print(i)
print()
# for와 range()를 사용하여 5부터 15까지 출력하시오.
for i in range(5, 16):
  print(i)
print()
# for와 range()를 사용하여 0~21까지의 짝수를 모두 출력하시오.
for i in range(22):
  if i % 2 == 0:
    print(i)
print()
# for와 range()를 사용하여 -100~100까지의 4의 배수를 모두 출력하시오.
for i in range(-100, 101, 4):
  print(i)
```

## 5. 연습문제

```python
# 문장을 입력받고 'q' 입력시 종료, 온점이 포함시 온점을 잘 입력하셨습니다.
sentence = input("문장을 입력해주세요. 'q' 입력시 종료합니다. ")
while True:
  if sentence == 'q' or sentence == 'Q':
    print("종료합니다.")
    break
  elif sentence[-1] == ".":
    print("온점을 잘 입력하셨습니다.")
  else:
    print("온점을 입력하세요.")
  sentence = input("문장을 입력해주세요. 'q' 입력시 종료합니다. ")
```

```python
# 1. width 변수에 인자가 두 개인 함수를 실행하여 두 숫자의 곱을 입력하여라.
def multiply(a ,b):
  return a * b


width = multiply(3, 5)
print(width)
# 2. width 변수에 (a + b) * c를 입력하는 함수 trepezium(a, b, c)를 만들고 실행하여라.
def trepezium(a, b, c):
  return (a + b) * c


width = trepezium(1, 2, 3)
print(width)
```
