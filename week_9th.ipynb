```python
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=["sepal length", "sepal width", "petal length", "petal width", "target"])
print(df.head(5))
```

```python
print(df.describe())
```

```python
print(df.groupby("target").size())
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["sepal length"], df["sepal width"])
df.hist(bins=20)
plt.grid()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis=1), df['target'])
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
clf = DecisionTreeClassifier(random_state=1)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)
print(pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
training_point=[[0.5, 0.2, 0.1], [0.9, 0.7, 0.3], [0.4, 0.5, 0.7]]
training_labels=[0, 1, 1]

clf.fit(training_point, training_labels)
unknown_point=[[0.2, 0.1, 0.7], [0.4, 0.7, 0.6], [0.5, 0.8, 0.1]]

pred = clf.predict(unknown_point)
print(pred)
```

```python
import pandas as pd
import numpy as np

breast_cancer = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data')
breast_cancer.head()
```

```python
breast_cancer.columns=["id_number","clump_thickness","unif_cell_size", "unif_cell_shape",
                       "marg_adhesion","single_epith_cell_size","bare_nuclei","bland_chromatin",
                       "normal_nucleli","mitoses","class"]
breast_cancer.head()
```

```python
breast_cancer.info()
breast_cancer.isnull().values.sum()
```

```python
# "?"   null 
breast_cancer['bare_nuclei']=breast_cancer['bare_nuclei'].replace("?", np.NAN)
breast_cancer['bare_nuclei']=breast_cancer['bare_nuclei'].fillna(breast_cancer['bare_nuclei'].value_counts().index[0])
breast_cancer['cancer_ind']=0
breast_cancer.loc[breast_cancer['class']==4, 'cancer_ind']=1

X=breast_cancer.drop(["id_number", "class", "cancer_ind"], axis=1)
y=breast_cancer.cancer_ind
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
y_pred = knn.predict(X_test_scaled)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(knn, {'n_neighbors':[1,2,3,4,5]}, n_jobs=-1, cv=7, scoring='roc_auc')
grid_search.fit(X_train_scaled, y_train)

grid_search.best_params_
knn_best = grid_search.best_estimator_
y_pred = knn_best.predict(X_test_scaled)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```
